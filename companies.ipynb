```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage= requests.get("https://www.ambitionbox.com/list-of-companies?search=CGI&page=1", headers=headers)
```

```python
soup= BeautifulSoup(webpage.text, 'lxml')
#print(soup.prettify())
```

```python
company= soup.find_all('div', class_="companyCardWrapper")
len(company)
```

```python

final=pd.DataFrame()

for j in range(1,500):
    
    webpage= requests.get("https://www.ambitionbox.com/list-of-companies?search=CGI&page=1", headers=headers)
    soup= BeautifulSoup(webpage.text, 'lxml')
    name=[]
    reviews=[]
    location=[]
    ctype=[]
    jobs=[]
    ratings=[]

    for i in company:
        name.append(i.find('h2').text.strip())
        ratings.append(i.find('div', class_="rating_text").text.strip())
        reviews.append(i.find('span', class_="companyCardWrapper__companyRatingCount").text.strip().replace("(", "").replace(")", ""))
        lnt= i.find('span', class_="companyCardWrapper__interLinking")
        temp= lnt.text.strip().split('|')
        ctype.append(temp[0])
        location.append(temp[1])
        jobs.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[3].text.strip())

    df=pd.DataFrame({'name':name,
    'rating':ratings,
    'reviews':reviews,
    'company_type':ctype,
    'Location': location,
    'Job_opening':jobs,
    })
  
    final=pd.concat([final,df],ignore_index=True)
    

```
