```python
m_list=[]
for i in range (1, 50):
    response = requests.get(f"https://api.themoviedb.org/3/movie/popular?language=en-US&page={i}", headers=headers)
    data= response.json()['results']
    m_list.extend(data)

df_combined= pd.DataFrame(m_list)

df2= df_combined[['id', 'title', 'original_language', 'vote_average', 'vote_count']]

df2.head()
```

```python
df2.to_csv('popular_movies.csv', index=False)
```
